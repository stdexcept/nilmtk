```python
import pandas as pd
import numpy as np
```

# Create some toy data

```python
PERIODS = 10
columns = [['meter1', 'meter1', 'meter1', 'meter2', 'meter2', 'meter2'],
           ['power', 'power', 'voltage', 'power', 'power', 'voltage'], 
           ['active', 'reactive', '', 'active', 'reactive', '']]
columns = pd.MultiIndex.from_arrays(columns)
rng = pd.date_range('2012', freq='S', periods=PERIODS)
df = pd.DataFrame(np.random.randint(low=0, high=1000, size=(10,6)),
                  index=rng,
                  columns=columns, dtype=np.float32)
df
```

## Save the data to a HDF5 file

```python
store = pd.HDFStore('store.h5', 'w', complevel=9, complib='blosc')
for meter in df.columns.levels[0]:
    store.put('building1/utility/electric/' + meter, df[meter])
store
store.close()
```

```python
class DataStore(object):
    def open(self, filename):
        self.store = pd.HDFStore(filename)
```
